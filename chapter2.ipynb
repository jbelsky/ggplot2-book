This is code related to [chapter 2](https://ggplot2-book.org/getting-started.html) of the book **ggplot2: Elegant Graphics for Data Analysis**.

```r
library("ggplot2")
```

```r
mpg
```

```r
# repr package installed with IRkernel provides ability to set plot parameter (like png())
options(repr.plot.width = 4, repr.plot.height = 3, repr.plot.res = 300)
```

```r
# Data is 'mpg', aesthetic mapping - engine size mapped to x position, fuel economy to y poistion, and points
ggplot(mpg, aes(x = displ, y = hwy)) + 
    geom_point()
```

```r
# Can add additional aesthetics such as 'color', 'shape', and 'size'
# ggplot2 automatically converts the data into aesthetics using a scale
ggplot(mpg, aes(x = displ, y = hwy)) + 
    geom_point(aes(color = class))
```

```r
# If set color outside of an aesthetic, then colors all the points the same color
ggplot(mpg, aes(displ, hwy)) + 
    geom_point(color = "blue")
```

```r
# data frame and mapping can be specified completely in geom_point()
ggplot() + 
    geom_point(data = mpg, mapping = aes(displ, hwy, color = class))
```

```r
# Faceting: trellis a graph by a certain variable (precede variable name by '~')
ggplot(mpg, aes(displ, hwy)) +
    geom_point() + 
    facet_wrap(~class)
```

```r
# geom_smooth() fits a smoother to the data and displays the smooth and its standard error
# Using method = 'loess' and formula 'y ~ x'
ggplot(mpg, aes(displ, hwy)) +
    geom_point() +
    geom_smooth()
```

```r
# Confidence interval display is controlled by 'se = TRUE' arg to 'geom_smooth()'
# Wiggliness of the line is controlled by the span paramter (0, exceedingly wiggly, to 1, not so wiggly)
ggplot(mpg, aes(displ, hwy)) +
    geom_point() +
    geom_smooth(se = F, span = 0.2)
```

```r
# 'method = "gam"' fits a generalized additive model provided by mgcv package
# Need to specify formula like 'formula = y ~ s(x)'
# ggplot2 uses 'formula = y ~ s(x, bs = "cs")' for large data (> 1,000 points)
library(mgcv)
ggplot(mpg, aes(displ, hwy)) +
    geom_point() +
    geom_smooth(method = "gam", formula = y ~ s(x))
```

```r
# 'method = "lm"' specifies a linear model
# 'method = "rlm"' uses a robust fitting algorithm (from MASS package) that is less affected by outliers
library(mgcv)
ggplot(mpg, aes(displ, hwy)) +
    geom_point() +
    geom_smooth(method = "lm")
```

## 2.6.2 Boxplots and jittered points

```r
# geom_jitter() adds random noise to help avoid overplotting
# 'width = 0.40' (default) is +/- percentage of implied bins (so 0.4 means that 80% of the implied bin will be occupied)
ggplot(mpg, aes(drv, hwy)) +
    geom_jitter(width = 0.2)
```

```r
# geom_boxplot()
ggplot(mpg, aes(drv, hwy)) +
    geom_boxplot(aes(color = drv), fill = "white", outlier.shape = 19)
```

```r
# geom_violin()
ggplot(mpg, aes(drv, hwy)) +
    geom_violin(aes(color = drv), fill = "white", bw = 1, draw_quantiles = 0.5, show.legend = F)
```
